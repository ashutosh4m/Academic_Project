```python
from pytube import YouTube
import os

def download(link, save_path):
    youtube_object = YouTube(link)
    youtube_object = youtube_object.streams.get_highest_resolution()

    try:
        youtube_object.download(output_path=save_path)
        print("Download is successful")
    except:
        print("Error downloading video")

# Input URL and desired save location from the user
link = input("Enter your URL: ")
save_location = input("Enter the directory where you want to save the video: ")

# Create the directory if it doesn't exist
if not os.path.exists(save_location):
    os.makedirs(save_location)

# Call the download function
download(link, save_location)
```

```python
from moviepy.editor import VideoFileClip

# Define the input video file and output audio file
video_file = "/content/drive/MyDrive/yardstick_assignment/Learn English Conversation - Bad News - 60 seconds (subtitles).mp4"
mp3_file = "/content/drive/MyDrive/yardstick_assignment/extracted_audio.mp3"

# Load the video clip
video_clip = VideoFileClip(video_file)

# Extract the audio from the video clip
audio_clip = video_clip.audio

# Write the audio to a separate file
audio_clip.write_audiofile(mp3_file)

# Close the video and audio clips
audio_clip.close()
video_clip.close()

print("Audio extraction successful!")
```

```python
from deepgram import (
    DeepgramClient,
    PrerecordedOptions,
    FileSource,
)

AUDIO_FILE = "/content/drive/MyDrive/yardstick_assignment/extracted_audio.mp3"


def main():
    try:
        deepgram = DeepgramClient("709b0b705b16b01839c0d23667a7f85887019f08")

        with open(AUDIO_FILE, "rb") as file:
            buffer_data = file.read()

        payload: FileSource = {
            "buffer": buffer_data,
        }

        options = PrerecordedOptions(
            model="nova-2",
            smart_format=True,
        )

        response = deepgram.listen.prerecorded.v("1").transcribe_file(payload, options)

        print(response.to_json(indent=4))

    except Exception as e:
        print(f"Exception: {e}")


if __name__ == "__main__":
    main()


```
